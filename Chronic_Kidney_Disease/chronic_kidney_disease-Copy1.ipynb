```python
df = pd.read_csv('chronic_kidney_disease.csv')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# rename of all feature name in broaded form
column_names = {"bp":"blood_pressure",
                "sg":"specific_gravity",
          "al":"albumin",
          "su":"sugar",
          "rbc":"red_blood_cells",
          "pc":"pus_cell",
          "pcc":"pus_cell_clumps",
          "ba":"bacteria",
          "bgr":"blood_glucose_random",
          "bu":"blood_urea",
          "sc":"serum_creatinine",
          "sod":"sodium",
          "pot":"potassium",
          "hemo":"haemoglobin",
          "pcv":"packed_cell_volume",
          "wc":"white_blood_cell_count",
          "rc":"red_blood_cell_count",
          "htn":"hypertension",
          "dm":"diabetes_mellitus",
          "cad":"coronary_artery_disease",
          "appet":"appetite",
          "pe":"pedal_edema",
          "ane":"anemia"
}

df.rename(columns = column_names, inplace = True)
```

```python
df
```

```python
df1 = df.drop('id', axis=1)
df1
```

```python
# check unique values of every feature
for i in df1:
    print(f'{i} : {df1[i].unique()}.\n')
```

```python
# substitution of wrong values with the specific correct values
df1['packed_cell_volume'] = df1['packed_cell_volume'].replace(to_replace = {'\t?':np.nan, '\t43':'43'})

df1['white_blood_cell_count'] = df1['white_blood_cell_count'].replace(to_replace = {'\t6200':'6200', '\t8400':'8400', '\t?':np.nan})

df1['red_blood_cell_count'] = df1['red_blood_cell_count'].replace(to_replace = '\t?', value = np.nan)

df1['diabetes_mellitus'] = df1['diabetes_mellitus'].replace(to_replace = {'\tno':'no', '\tyes':'yes', ' yes':'yes'})

df1['coronary_artery_disease'] = df1['coronary_artery_disease'].replace(to_replace = '\tno', value = 'no')



df1['classification'] = df1['classification'].replace(to_replace = 'ckd\t', value = 'ckd')
```

```python
for i in df1:
    print(f'{i} : {df1[i].unique()}.\n')
```

```python
df1.info()
```

```python
# the three below columns are in dataset as object class, so make them numeric in the below
df1['packed_cell_volume'] = pd.to_numeric(df1['packed_cell_volume'])

df1['white_blood_cell_count'] = pd.to_numeric(df1['white_blood_cell_count'])

df1['red_blood_cell_count'] = pd.to_numeric(df1['red_blood_cell_count'])
```

```python
df1.info()
```

```python
df1.isnull().sum() 
```

```python
df1.isnull().sum().sum()
```

```python
df1
```

```python
# numerical_columns_with_missing_values
numerical_columns_with_missing_values = df1[df1.select_dtypes(include=['float64', 'int64']).columns].isnull().sum()
numerical_columns_with_missing_values
```

```python
# make list of the index names
numerical_columns_with_missing_values_list = numerical_columns_with_missing_values.index.tolist()
numerical_columns_with_missing_values_list
```

```python
# fill the numerical missing values with iterative imputer
imputer = IterativeImputer()
imputed_numerical_data = imputer.fit_transform(df1[numerical_columns_with_missing_values_list])
```

```python
imputed_numerical_df1 = pd.DataFrame(imputed_numerical_data, columns = numerical_columns_with_missing_values_list)
imputed_numerical_df1
```

```python
#categorical_columns = df1.select_dtypes(include=['object']).columns
categorical_columns_with_missing_values = df1[df1.select_dtypes(include=['object']).columns].isnull().sum()
categorical_columns_with_missing_values
```

```python
categorical_columns_with_missing_value_list = categorical_columns_with_missing_values.drop('classification').index.tolist()
categorical_columns_with_missing_value_list
```

```python
# fill missing value in categorical columns with forward fill
ffilled_categorical_df1 = df1[categorical_columns_with_missing_value_list].fillna(method='bfill')
ffilled_categorical_df1
```

```python
# checking the missing value after forward fill
ffilled_categorical_df1.isnull().sum()
```

```python
# unique value counts of 'pus_cell' column
ffilled_categorical_df1['pus_cell'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
# encode the categorical data in numerical data with LabelEncoder
label_enco = LabelEncoder()
encoded_categorical_data = ffilled_categorical_df1.apply(label_enco.fit_transform)
encoded_categorical_data
```

```python
# combine numerical and categorical data
combined_features_df = pd.concat([imputed_numerical_df1, encoded_categorical_data], axis=1)
```

```python
combined_features_df
```

```python
df1['classification'].value_counts()
```

```python
# separate target column from df1
target = df1['classification']
target
```

```python
# add target column in the combined_features_df dataset
combined_features_df['classification'] = target
combined_features_df
```

```python
from sklearn.model_selection import train_test_split

X = combined_features_df.drop('classification', axis=1)
y = combined_features_df['classification']

# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
X_train
```

```python
y_train
```

```python
from sklearn.svm import SVC
svm_model = SVC(kernel='linear', C=10, gamma='auto')
```

```python
# fit the svm algorithm with train data
svm_model.fit(X_train, y_train)
```

```python
# find model's score with test data
svm_model_accuracy = svm_model.score(X_test, y_test)
```

```python
print(f'Model accuracy {svm_model_accuracy}')
```

```python
svm_model_accuracy_percentage = svm_model_accuracy * 100
print(f'Model accuracy in percentage {svm_model_accuracy_percentage:.2f}%')
```
